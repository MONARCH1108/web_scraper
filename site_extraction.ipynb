```python
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get('https://en.wikipedia.org/wiki/Cyrenaics')
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
paragraph = soup.find('div', class_='mw-body-content')
p = paragraph.find_all('p')
with open('Cyrenaicism.text','w',encoding='UTF-8') as f:
    for para in p:
        text = para.get_text(separator=' ', strip=True)
        print(text)
        f.write(text + '\n')
```
